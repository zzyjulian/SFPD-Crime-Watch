When going through some of the data, I realized that the time listed on "incident time column " tends to jump all over the place meaning it is really inconsistant. 

For example, there are a few rows that tends to have reports from 2025, but then suddenly jump to 2024 and then back to 2025. 

During the same time, I also noticed that the mounth and date tends to have the same problem. 

What I did below is to reorganized this column into a decdending time order starting with nearist date time in 2025 than all the way back to 2018.

```python
crimes['Incident Datetime'] = pd.to_datetime(crimes['Incident Datetime'], errors='coerce')
```

```python
crimes = crimes.sort_values('Incident Datetime', ascending=False)
```

```python
crimes = crimes.reset_index(drop=True)
```

```python
crimes.to_csv('crimes_sorted.csv', index=False)
```
